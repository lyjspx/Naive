```python
import xlrd
```

```python
paper2004=xlrd.open_workbook("Kalia_et_al-2018.xlsx","r")
```

```python
sheet1A=paper2004.sheet_by_index(0)
```

```python
allMarker=[]
for sheet in paper2004.sheets():
    marker=[]
    for row in range(sheet.nrows):
        thisRow=[]
        for col in range(sheet.ncols):
            if(sheet.cell(row,col).value !=''):
                thisRow.append(sheet.cell(row,col).value)
        marker.append(thisRow)
    for one in marker:
        if(len(one)==2):
            allMarker.append([one[0],one[1],sheet.name])
        elif(len(one)>2):
            for i in range(len(one)-1):
                allMarker.append([one[i],one[len(one)-1],sheet.name])
```

```python
allMarker
```

```python
with open("Kalia_et_al_2018","w") as outputf:
    for line in allMarker:
        outputf.write(",".join([line[0],str(line[1]),line[2]])+"\n")
```
