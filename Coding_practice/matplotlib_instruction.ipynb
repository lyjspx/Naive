### Single line plot

```python
import random
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
plt.figure(figsize = (12,6))
plt.plot(x,y_shanghai)

#create tick
x_label = ["11:{}".format(i) for i in x]
plt.yticks(range(10,30)[::5])
plt.xticks(x[::5],x_label[::5])

#creat grid
plt.grid(True,linestyle="--",alpha = 0.5)

#create labels and title
plt.xlabel("Temp")
plt.ylabel("city")
plt.title("I am a figure!")

plt.show() 
```

### Multiple line 

```python

x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(25,28) for i in x]
plt.figure(figsize = (12,6))
plt.plot(x,y_shanghai,color = "r", linestyle = "--",label = "city1")
plt.plot(x,y_beijing, color = "b", label = "city2")

#create legend
plt.legend(loc = "lower right")

#create tick
x_label = ["11:{}".format(i) for i in x]
plt.yticks(range(10,30)[::5])
plt.xticks(x[::5],x_label[::5])

#creat grid
plt.grid(True,linestyle="--",alpha = 0.5)

#create labels and title
plt.xlabel("Temp")
plt.ylabel("city")
plt.title("I am a figure!")

plt.show() 
```

# Multiple plots

Single plot is procedure oriented;  
Multiple plots are object oriented;


```python
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(25,28) for i in x]

#create canvas
figure, axes= plt.subplots(nrows=1, ncols=2,figsize=(18,6),dpi=60)

axes[0].plot(x,y_shanghai,color = "r", linestyle = "--",label = "city1")
axes[1].plot(x,y_beijing, color = "b", label = "city2")

#create legend
axes[0].legend(loc = "lower right")
axes[1].legend(loc = "lower right")

#create tick ##### Note:different in single plot
x_label = ["11:{}".format(i) for i in x]
axes[0].set_yticks(range(10,30)[::5])
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label[::5])
axes[1].set_yticks(range(10,30)[::5])
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_label[::5])


    ###Multiple lines editing: Mac(cmd + click)

#creat grid
axes[0].grid(True,linestyle="--",alpha = 0.5)
axes[1].grid(True,linestyle="--",alpha = 0.5)

#create labels and title
axes[0].set_xlabel("Temp")
axes[0].set_ylabel("city")
axes[0].set_title("I am a figure!")
axes[1].set_xlabel("Temp")
axes[1].set_ylabel("city")
axes[1].set_title("I am a figure 2!")

plt.show() 
```

### Bar plot

```python
x = ["a","b","c","d","f","i"]
y = [random.randrange(0,100) for i in x]
```

```python
plt.figure(figsize=(8,4),dpi=80)

x_ticks = range(len(x))

plt.bar(x_ticks,y) ###add color parameter here

plt.xticks(x_ticks,x)

plt.grid(alpha=0.2)

plt.show()
```

```python
#Shift bar pos and ticks pos

plt.figure(figsize=(8,4),dpi=80)

x_ticks = range(len(x))

plt.bar(x_ticks[0:2],y[0:2],width=0.2)
plt.bar([0.2,1.2],y[2:4],width=0.2)

plt.legend(x[0:2])

plt.xticks([0.1,1.1],x)

plt.grid(alpha=0.2)

plt.show()
```

### Histogram

```python
x = range(0,20)
y = [int(random.gauss(50, 20)) for i in x]
```

```python
plt.figure(figsize=(10,5))

distance = 20

group_number = int(max(y)-min(y)/distance)

plt.hist(y,bins=group_number)

plt.xticks(range(min(y),max(y)+3,distance))

plt.show()
```
