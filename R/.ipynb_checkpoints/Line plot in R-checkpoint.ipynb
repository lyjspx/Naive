```r
marker = read.csv("/Users/desert/OneDrive - North Dakota University System/Alignment paper revision/Marker_number.csv",
                  header = T,row.names = 1)
```

#### Cheating sheet in plot() in R

|keyword|functions|
|:--------:|:----------:|
|xlim/ylim|axis range|
|axes|plot axes or not|
|xlab/ylab|axis title|
|cex|text size|
|type|line, dot or both|
|pos|postion of text, 4 options|
|pch|shape of data point|
|at/label|corresponding to each|
|superscript|expression(paste("1e"^4))|
|bty|frame|

```r
plot(marker[,1],type = "b",xlim = c(1,10),ylim = c(-20000,800000),axes = F
     ,xlab = "Maximum missing percentage",ylab = "Number of Markers")
text(marker[,1]+10000,labels = marker[,1],pos = 3,cex = 0.6)
lines(marker[,2],type = "b",pch = 2)  
text(marker[,2],labels = marker[,2],pos = 1,cex = 0.6)
axis(1,at=1:10,labels = c("All","90%","80%","70%","60%","50%","40%","30%","20%","10%"),tick = T)
axis(2,las= 1,at=c(10000,100000,300000,500000,800000),
     labels = c(expression(paste("1e"^4)),expression(paste("1e"^5)),expression(paste("3e"^5)),expression(paste("5e"^5)),expression(paste("8e"^5)))
     ,cex=0.8)
legend(7,700000,legend = c("Enrichment","Random"),pch = 1:2,cex = 0.8,bty = "n")
```
